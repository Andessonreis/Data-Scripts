```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
uploaded_file = next(iter(uploaded))
data = pd.read_csv(uploaded_file)
```

```python
def basic_statistics(data):
    """Calculate basic statistics for selected columns."""
    statistics = data.describe()
    return statistics


def scatter_plot(data, x_col, y_col):
    """Create a scatter plot between(dispersão) two columns."""
    plt.figure(figsize=(10, 6))
    plt.scatter(data[x_col], data[y_col], alpha=0.5)
    plt.title(f'Scatter Plot: {x_col} vs {y_col}')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.grid(True)
    plt.show()

# Calcular estatísticas básicas
selected_columns = ['pl_orbper', 'pl_rade', 'pl_bmasse']
statistics = basic_statistics(data[selected_columns])

# Exibindo estatísticas básicas
print(statistics)

# Exibindo um gráfico de dispersão entre pl_orbper e pl_rade
scatter_plot(data, 'pl_orbper', 'pl_rade')
```
